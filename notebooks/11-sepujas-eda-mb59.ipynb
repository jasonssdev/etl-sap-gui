```python
import pandas as pd
import os
```

```python
current_path = os.getcwd()
print(current_path)
```

```python
base_path = os.path.abspath(os.path.join(current_path, os.pardir))
print(base_path)
```

```python
mb59_file_path = os.path.join(base_path, 'data', 'raw', 'tbl_mb59.txt')
print(mb59_file_path)
```

```python
mb59_preprocessed_path = os.path.join(base_path, 'data', 'preprocessed', 'tbl_mb59.csv')
print(mb59_preprocessed_path)
```

```python
df_mb59 = pd.read_csv(mb59_file_path, sep='\t', skiprows=3, encoding='latin1')
```

```python
df_mb59.head()
```

```python
unnamed_columns = [col for col in df_mb59.columns if 'Unnamed:' in col]
df_mb59.drop(columns=unnamed_columns, inplace=True)
column_titles = df_mb59.columns.tolist()
new_column_titles = {col: col.strip().replace(' ', '_').replace('-','_').replace('.','') for col in column_titles}
df_mb59.rename(columns=new_column_titles, inplace=True)

```

```python
column_counts = df_mb59.columns.value_counts()
duplicate_columns = column_counts[column_counts > 1].index

for col in duplicate_columns:
    col_indices = [i for i, x in enumerate(df_mb59.columns) if x == col]
    for j, index in enumerate(col_indices):
        df_mb59.columns.values[index] = f"{col}_{j+1}"
```

```python
def transform_columns(df):
    str_columns = [
        'Qty_in_UnE', 'EUn', 'Material_Description', 'Movement_Type_Text', 'Rec', 'Mvt', 'TETy'
    ]
    for col in str_columns:
        df[col] = df[col].astype(str).str.strip()

    numeric_columns = ['CoCd', 'Plnt', 'Material', 'SLoc', 'MvT', 'Mat_Doc', 'Item', 'PO','Supplier', 'Counter']
    for col in numeric_columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int).astype(str)

    df['Amount_in_LC'] = df['Amount_in_LC'].astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
    df['Quantity'] = df['Quantity'].astype(str).str.replace('.', '').str.replace(',', '.').astype(float)


    date_columns = ['Pstng_Date']
    for col in date_columns:
        df[col] = pd.to_datetime(df[col], errors="coerce", format='%d.%m.%Y')

transform_columns(df_mb59)
```

```python
df_mb59['key_plant'] = df_mb59['Plnt'] + '/' + df_mb59['Material']
df_mb59['key_plant'] = df_mb59['key_plant'].astype(str).str.strip()
```

```python
df_mb59.to_csv(mb59_preprocessed_path, index=False)
```
