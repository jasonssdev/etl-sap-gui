```python
import pandas as pd
import os
```

```python
current_path = os.getcwd()
print(current_path)
```

```python
base_path = os.path.abspath(os.path.join(current_path, os.pardir))
print(base_path)
```

```python
bo2mm_file_path = os.path.join(base_path, 'data', 'raw', 'tbl_bo2mm.txt')
print(bo2mm_file_path)
```

```python
bo2mm_preprocessed_path = os.path.join(base_path, 'data', 'preprocessed', 'tbl_bo2mm.csv')
print(bo2mm_preprocessed_path)
```

```python
df_bo2mm = pd.read_csv(bo2mm_file_path, sep='\t', skiprows=3, encoding='latin1')
```

```python
df_bo2mm.head()
```

```python
unnamed_columns = [col for col in df_bo2mm.columns if 'Unnamed:' in col]
```

```python
df_bo2mm.drop(columns=unnamed_columns, inplace=True)
```

```python
column_titles = df_bo2mm.columns.tolist()
```

```python
new_column_titles = {col: col.strip().replace(' ', '_').replace('-','_').replace('.','') for col in column_titles}
```

```python
df_bo2mm.rename(columns=new_column_titles, inplace=True)
```

```python
column_counts = df_bo2mm.columns.value_counts()
duplicate_columns = column_counts[column_counts > 1].index

for col in duplicate_columns:
    col_indices = [i for i, x in enumerate(df_bo2mm.columns) if x == col]
    for j, index in enumerate(col_indices):
        df_bo2mm.columns.values[index] = f"{col}_{j+1}"
```

```python
df_bo2mm.columns.tolist()
```

```python
df_bo2mm.dtypes
```

```python
df_bo2mm.head()
```

```python
df_bo2mm['Material'] = df_bo2mm['Material'].fillna(0).astype(int).astype(str).str.strip()
df_bo2mm['Material_Description'] = df_bo2mm['Material_Description'].astype(str).str.strip()
df_bo2mm['Plnt'] = df_bo2mm['Plnt'].fillna(0).astype(int).astype(str)
df_bo2mm['MS'] = df_bo2mm['MS'].fillna(0).astype(int).astype(str)
df_bo2mm['Typ'] = df_bo2mm['Typ'].astype(str).str.strip()
df_bo2mm['Combo_Ind'] = df_bo2mm['Combo_Ind'].astype(str).str.strip()
df_bo2mm['BO_items'] = df_bo2mm['BO_items'].fillna(0).astype(int).astype(str).str.strip()
df_bo2mm['GrpValue'] = df_bo2mm['GrpValue'].astype(str).str.strip().str.replace('.', '').str.replace(',', '.').astype(float)
df_bo2mm['Curr'] = df_bo2mm['Curr'].astype(str).str.strip()
df_bo2mm['BO_quan'] = df_bo2mm['BO_quan'].astype(str).str.strip().str.replace('.', '').str.replace(',', '.').astype(float)
df_bo2mm['Avail_QTY'] = df_bo2mm['Avail_QTY'].astype(str).str.strip().str.replace('.', '').str.replace(',', '.').astype(float)
df_bo2mm['AltATP'] = df_bo2mm['AltATP'].fillna(0).astype(int).astype(str).str.strip()
df_bo2mm['SU'] = df_bo2mm['SU'].astype(str).str.strip()
df_bo2mm['Altern?'] = df_bo2mm['Altern?'].fillna(0).astype(int).astype(str).str.strip()
df_bo2mm['MM_note'] = df_bo2mm['MM_note'].astype(str).str.strip()
df_bo2mm['BO_Note'] = df_bo2mm['BO_Note'].astype(str).str.strip()
df_bo2mm['FocCust'] = df_bo2mm['FocCust'].astype(str).str.strip()
df_bo2mm['lpostp'] = pd.to_datetime(df_bo2mm['lpostp'], errors="coerce", format='%d.%m.%Y')
df_bo2mm['maxDlvDate'] = pd.to_datetime(df_bo2mm['maxDlvDate'], errors="coerce", format='%d.%m.%Y')
df_bo2mm['ResCount'] = df_bo2mm['ResCount'].fillna(0).astype(int).astype(str).str.strip()
df_bo2mm['maxiPDdel'] = df_bo2mm['maxiPDdel'].fillna(0).astype(int).astype(str).str.strip()
df_bo2mm['maxRDdelay'] = df_bo2mm['maxRDdelay'].fillna(0).astype(int).astype(str).str.strip()
df_bo2mm['maxPDdelay'] = df_bo2mm['maxPDdelay'].fillna(0).astype(int).astype(str).str.strip()
df_bo2mm['RES_QTY'] = df_bo2mm['RES_QTY'].astype(str).str.strip()
df_bo2mm['PO_qty'] = df_bo2mm['PO_qty'].astype(str).str.strip().str.replace('.', '').str.replace(',', '.').astype(float)
df_bo2mm['Next_PO'] = df_bo2mm['Next_PO'].fillna(0).astype(int).astype(str).str.strip()
df_bo2mm['PO_qty1'] = df_bo2mm['PO_qty1'].astype(str).str.strip().str.replace('.', '').str.replace(',', '.').astype(float)
df_bo2mm['Delivdate'] = pd.to_datetime(df_bo2mm['Delivdate'], errors="coerce", format='%d.%m.%Y')
df_bo2mm['TABCD'] = df_bo2mm['TABCD'].astype(str).str.strip()
df_bo2mm['XYZ'] = df_bo2mm['XYZ'].astype(str).str.strip()
df_bo2mm['PDT'] = df_bo2mm['PDT'].fillna(0).astype(int).astype(str).str.strip()
df_bo2mm['ProcType'] = df_bo2mm['ProcType'].astype(str).str.strip()
df_bo2mm['SPT'] = df_bo2mm['SPT'].astype(str).str.strip()
df_bo2mm['lastResc'] = pd.to_datetime(df_bo2mm['lastResc'], errors="coerce", format='%d.%m.%Y')
df_bo2mm['PDcount'] = df_bo2mm['PDcount'].fillna(0).astype(int).astype(str).str.strip()
df_bo2mm['MRPCn'] = df_bo2mm['MRPCn'].astype(str).str.strip()
df_bo2mm['Name'] = df_bo2mm['Name'].astype(str).str.strip()

```

```python
df_bo2mm['key_plant'] = df_bo2mm['Plnt'] + '/' + df_bo2mm['Material']
df_bo2mm['key_plant'] = df_bo2mm['key_plant'].astype(str).str.strip()
```

```python
df_bo2mm.to_csv(bo2mm_preprocessed_path, index=False, sep='|')
```
