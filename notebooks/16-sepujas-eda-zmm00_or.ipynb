```python
import pandas as pd
import os
```

```python
current_path = os.getcwd()
print(current_path)
```

```python
base_path = os.path.abspath(os.path.join(current_path, os.pardir))
print(base_path)
```

```python
zmm00_or_file_path = os.path.join(base_path, 'data', 'raw', 'tbl_zmm00_or.txt')
print(zmm00_or_file_path)
zmm00_or_preprocessed_path = os.path.join(base_path, 'data', 'preprocessed', 'tbl_zmm00_or_file_path.csv')
print(zmm00_or_preprocessed_path)
```

```python
df_zmm00_or = pd.read_csv(zmm00_or_file_path, sep='\t', skiprows=3, encoding='latin1')
```

```python
df_zmm00_or.head()
```

```python
unnamed_columns = [col for col in df_zmm00_or.columns if 'Unnamed:' in col]
df_zmm00_or.drop(columns=unnamed_columns, inplace=True)
column_titles = df_zmm00_or.columns.tolist()
new_column_titles = {col: col.strip().replace(' ', '_').replace('-','_').replace('.','') for col in column_titles}
df_zmm00_or.rename(columns=new_column_titles, inplace=True)
```

```python
column_counts = df_zmm00_or.columns.value_counts()
duplicate_columns = column_counts[column_counts > 1].index
```

```python
for col in duplicate_columns:
    col_indices = [i for i, x in enumerate(df_zmm00_or.columns) if x == col]
    for j, index in enumerate(col_indices):
        df_zmm00_or.columns.values[index] = f"{col}_{j+1}"
```

```python
def transform_columns(df):
    str_columns = [
        'PGr', 'Material_Description', 'Confirmed', 'Descript', 'ItCa', 'Status', 'Reference', 'ActMatAvDt', 'Conf_Qty', 'Ship_to', 'MRPCD']
    for col in str_columns:
        df[col] = df[col].astype(str).str.strip()

    numeric_columns = ['Region', 'Plnt', 'Descriptn', 'Reservno', 'Itm', 'Material', 'Ship_Code', 'Diff_Days', 'Ext_OrdNo', 'Ext_OrdPo', 'Plnt1', 'Pur_Doc', 'Item', 'AlrtExists', 'DDNP', 'DSLU', 'DDGT', 'NOC', 'WRGT', 'AlrtSum', 'Goods_Supp', 'Name', 'Customer', 'Name1', 'MRPCS', 'Name2', 'D797_uCnt', 'CommentBU', 'CommentMO']
    for col in numeric_columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int).astype(str)

    df['Reqmnt_qty'] = df['Reqmnt_qty'].astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
    df['Quantity'] = df['Quantity'].astype(str).str.replace('.', '').str.replace(',', '.').astype(float)


    date_columns = ['Base_date', 'Orig_Date', 'ReqmtsDate', 'Created_on', 'Conf_Date', 'D797_date']
    for col in date_columns:
        df[col] = pd.to_datetime(df[col], errors="coerce", format='%d.%m.%Y')

transform_columns(df_zmm00_or)
```
